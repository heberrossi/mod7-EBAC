#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável.
    Dica:
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis,
        3. conte o número de categorias dessa variável
        4. acumule essa informação de 3. na lista que você criou em 1.
        5. No final, essa lista pode ser atribuída à nossa variável.
3. Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}.

````python
metadados = pd.DataFrame({'Nome da Variável': df.columns, 'Tipo de Dados': df.dtypes})
metadados.reset_index(drop=True, inplace=True)
````

````python
qtd_categorias = []

# Loop para calcular o número de categorias de cada variável
for col in df.columns:
    qtd_categorias.append(df[col].nunique())

# Cria uma nova coluna no DataFrame pela lista
metadados['qtd_categorias'] = qtd_categorias

print(metadados)
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
metadado_final = pd.get_dummies(df, columns=['sexo', 'posse_de_veiculo',
                                        'posse_de_imovel', 'tipo_renda',
                                        'educacao', 'estado_civil',
                                        'tipo_residencia'], drop_first=True)
metadado_final
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
metadado_final['possui_email'].value_counts()
````

````python
metadado_final['posse_de_veiculo_Y'].value_counts()
````

````python
metadado_final['mau'].value_counts()
````

````python
tabela_cruzada = pd.crosstab(metadado_final['possui_email'], metadado_final['posse_de_veiculo_Y'])
tabela_cruzada
````

````python
tabela_cruzada_proporcao = pd.crosstab(metadado_final['possui_email'],
                                       metadado_final['posse_de_veiculo_Y'],
                                       normalize='index')*100
tabela_cruzada_proporcao
````

````python
tabela_cruzada2 = pd.crosstab(index=metadado_final['possui_email'],
                              columns=metadado_final['posse_de_veiculo_Y'],
                              values=metadado_final['mau'],
                              aggfunc='sum',
                              margins=True
                              )

tabela_cruzada2.columns = ['Não Possui Veículo', 'Possui Veículo', 'Total']

tabela_cruzada2.index = ['Não Possui Email', 'Possui Email', 'Total']

print(tabela_cruzada2)
````
